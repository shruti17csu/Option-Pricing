```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
def stimulate_geometric_random_walk(s0,T=2,N=1000,mu=0.1,sigma=0.05):#s0 intial value of stock
    #T= expiry
    dt=T/N
    t=np.linspace(0,T,N)
    # standard normal distribution mean=0 std=1
    W=np.random.standard_normal(size=N)# this is dW 
    # N(0,dt)=sqrt(dt)*N(0,1)
    W=np.cumsum(W)*np.sqrt(dt)#W(t)=W(t-1)+dW
    X=(mu-0.5*sigma**2)*t+sigma*W
    s=s0*np.exp(X)
    return t,s
def plot_result(t,s):
    plt.plot(t,s)
    plt.xlabel("time d(t)")
    plt.ylabel("stock price s(t)")
    plt.title("Geometric Brownian Motion")
    plt.show()
```

```python
if __name__=="__main__":
    time ,data=stimulate_geometric_random_walk(10)
    plot_result(time,data)
```
